1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 10*x - 5

delta_x = 0.1
def der_f(x):
    return (f(x + delta_x) -f(x))/delta_x

x = np.linspace(0, 10, 10)
plt.plot(x, f(x), 'o', label='function', color='blue')
plt.plot(x, der_f(x), 'o',label= 'derivative', color='orange')
plt.title('Function and derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')
f1 = sp.sin(x)
df1 = sp.diff(f1)
print(f1, df1)

x = np.linspace(0, 10, 100)
f1 = np.sin(x)
df1 = np.gradient(f1, x)
plt.plot(x, f1, label='function', color='blue')
plt.plot(x, df1, label='derivative', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = sp.symbols('x')
f2 = sp.ln(x)
df2 = sp.diff(f2)
print(f2, df2)

x = np.linspace(0.1, 10, 100)
f2 = np.log(x)
df2 = np.gradient(f2, x)
plt.plot(x, f2, label='function', color='blue')
plt.plot(x, df2, label='derivative', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
def f(x):
    return x**2 -5*x + 1

x = sp.symbols('x')
df = sp.diff(f(x))
print(df)

x = np.linspace(0, 100, 100)
df = np.gradient(f(x), x)

plt.plot(x, f(x), label='function', color='blue')
plt.plot(x, df, label='derivative', color='orange')
plt.xlabel('Manufactered parts')
plt.ylabel('Total costs of the company')
plt.legend()
plt.show()
```
